```python
# Example 2 - print 1 to 5
i = 1
while i <= 5:
    print(i)
    i += 1
```

```python
# Example 3 - print 5 to 1
i = 5
while i > 0:
    print(i)    
    i -= 1
```

```python
# Multiplication table of N
n = int(input("enter n: "))
i = 1

while i <= 10:
    print(i * n)     
    i += 1
```

```python
# Break & Continue

# Break for multiple of 6
i = 1

while i <= 10:
    if(i % 6 == 0):
        break
    print(i)           # 1, 2, 3, 4, 5, break
    i += 1
```

```python
# Skip multiples of 3
i = 0

while(i < 10):
    i += 1
    if(i % 3 == 0):
        continue;      # 1, 2, continue, 4, 5, continue, 7, 8, continue, 10
    print(i)
```

```python
# Print odd nums from 1 to 10 using continue
i = 0

while(i < 10):
    i += 1
    if(i % 2 == 0):
        continue;      # 1, 3, 5, 7, 9
    print(i)
```

**FOR LOOP**

```python
# for Loop - Example 1

for i in range(5):     # 0, 1, 2, 3, 4
    print(i)
```

```python
# Chars of a string
word = "Prime"

for ch in word:
    print(ch)

# Check if char 'i' exists in word
if 'i' in word:
    print("letter exists")
```

```python
# Example 2 - count number of i's in word

word = "artificial intelligence"
count = 0

for ch in word:
    if ch == 'i':
        count += 1

print(f"i occurs {count} times.")
```

```python
# Example 3 - count vowels in word
for ch in word:
    if (ch == 'a' or ch == 'a' or ch == 'a' or ch == 'a' or ch == 'a'):
        count += 1

print(f"vowel count = {count}")
```

```python
# Sum of first n natural nums

n = int(input("enter n: "))
sum = 0
for i in range(1, n+1):
    sum += i

print("sum = ", sum)
```
