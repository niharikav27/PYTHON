```python
# wap which takes string as input and replace every occurence of string in a given list to hii 
# and alsp print list which have length of all string inside that list

string = "am"
lst = ['I','am','programmer','and','I','am','coder']

new_lst =[s.replace(string,"hii") for s in lst]
length_lst = [len(s) for s in new_lst]

print(new_lst)
print(length_lst)
```

```python
# write a program to remove all vowels from a string

string = input("Enter a string : ")
result = ""
for ch in string:
    if ch.lower() not in "aeiou":
        result += ch

print(result)    
```

```python
#wap which takes character as a input and count the occurence of that character into the given string
# and convert all them into '#'

string = input("Enter a string: ")
ch = input("Enter a character: ")

count = string.count(ch)
result = string.replace(ch, '#')

print(count)
print( result)
```

```python
#  wap that splits a given string into words , sort the words in alphabetical order 
# and joins them back into a single string

string = input("Enter a string: ")

words = string.split()

words.sort()

result = " ".join(words)

print(result)
```

```python
#wap which takes string as input and swap first and last character of that string

string = input("Enter a string: ")

lst = list(string)
lst[0], lst[-1] = lst[-1], lst[0]

print( "".join(lst))
```

```python
#wap to check if a given string is a pallindrome

string = input("Enter a string: ")      #racecar
                  
if string == string[::-1]:
    print("Palindrome")
else:
    print("Not a Palindrome")
```

```python
#ask the user for a string and print 
# all unique charcters
# the count of unique characters


# first approach to use set and get unordered output
string = input("Enter a string: ")

unique_chars = set(string)

print("Unique characters:", unique_chars)
print("Count of unique characters:", len(unique_chars))


# if we want ordered output
string = input("Enter a string: ")

unique = ""

for ch in string:
    if ch not in unique:
        unique += ch

print("Unique characters:", unique)
print("Count:", len(unique))
```
