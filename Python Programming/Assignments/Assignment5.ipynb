```python
# wap which takes string as input and replace every occurence of string in a given list to hii 
# and alsp print list which have length of all string inside that list

string = "am"
lst = ['I','am','programmer','and','I','am','coder']

new_lst =[s.replace(string,"hii") for s in lst]
length_lst = [len(s) for s in new_lst]

print(new_lst)
print(length_lst)
```
