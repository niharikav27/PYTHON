```python
#wap to create dictionary naed student_grades with following key value pair: Alice:85 ,Bob:90 ,Charlie:78
#also print the dictionary

student_grades = {'Alice':85 , 'Bob':90 , 'Charlie':78}
print(student_grades)
```

```python
#wap to acccess and print the grade of bob from the above dictionary
bob_grades = student_grades.get('Bob')
print(bob_grades)
```

```python
#wap to add new student 'David' with a grade of 92 to the above dictionary and print it
student_grades['David'] = 92
print(student_grades)
```

```python
#wap to update the grade of charlie to 80 in the above dictionary and print it

student_grades['Charlie'] = 80
print(student_grades)
```

```python
#wap to delete the student David fro the given dictionary using the del keyword ,also print the updated dictionary

del student_grades['David']
print(student_grades)
```

```python
#wap to check if the key 'Eve' is present in the given dictionary if not then print "Eve is not present in the dictionary"

if 'Eve' in student_grades:
    print("Eve is present in the dictionary")
else:
    print("Eve is not present in the dictionary")    
```

```python
#wap to print all the student names in the given dictionary

for k in student_grades.keys():
    print(k)
```

```python
#wap to print all the grades in the given dictionary

for v in student_grades.values():
    print(v)
```

```python
#wap to print all key-values pairs in the given dictionary in the format "student:grade"

for i in student_grades.items():
    print(i)
```

```python
#wap to clear all elements from the given dictionary

student_grades.clear()
print(student_grades)
```

```python
#wap to check if a dictionary is empty or not

if not student_grades:
    print("Dictionary is empty")
else:
    print("Dictionary is not empty")

#another way
if len(student_grades) == 0:
    print("Dictionary is empty")
```

```python
#wap to count how many key value pairs are present in the dictionary by traversing on the given dictionary

student_grades = {'Alice':85 , 'Bob':90 , 'Charlie':78}
count = 0

for k in student_grades:
    count += 1

print(count)
```

```python
#wap to access dictionary keys element by index(user input)
index = int(input("Enter index (starting from 1): "))

keys_list = list(student_grades.keys())

if 1 <= index <= len(keys_list):
    print("Key at index:", keys_list[index - 1])
else:
    print("Invalid index")
```
