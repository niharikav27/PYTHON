```python
#Write a program that takes salary as input .using conditional statements, calculate tge final tax rate based on these rules:
# if salary < 30,000 -> 5%
# if salary 30,000 - 70,000 -> 15%
# if salary > 70,000 -> 25%

salary = int(input("Enter the salary : "))
if( salary < 30000):
    final_tax_rate = (salary * 5)/100
elif(salary >= 30000 and salary <= 70000):
    final_tax_rate = (salary * 15)/100  
else:
    final_tax_rate = (salary * 25)/100 
    
print(final_tax_rate)         
```

```python
# wap that takes 4 angles as input and checks whwther it can form quadrilateral or not

a1 = int(input("Enter first angle : "))
a2 = int(input("Enter second angle : "))
a3 = int(input("Enter third angle : "))
a4 = int(input("Enter fourth angle : "))

sum = a1 + a2 + a3 + a4

if(sum == 360):
    print("quadrilateral")
else:
    print("not a quadrilateral")    
```
