```python
# write a function to print even numbers from a to b (inclusive)

def even(a,b):
    for i in range(a , b+1):
        if i % 2 == 0 :
            print( i, end = " ")

even(2,10)
```

```python
# wap that prints digit of a number

def print_digits(n):
    while n > 0:
        digit = n % 10
        print(digit)
        n = n // 10

print_digits(312)
        
```

```python
#wap to print count of digits in  a number

def print_count(n):
    count = 0;
    while n > 0:
        digit = n % 10
        count += 1
        n = n // 10
        
    return count
print_count(312)
```

```python
#wap to return the sum of digits of a number

def print_sum(n):
    sum = 0
    while n>0 :
        digit = n % 10
        sum += digit
        n = n // 10
    return sum
print_sum(14)    
```

```python
# create a calculator that performs arithmetic operations.create a function 
# calculator(a,b,opertaion) that performs + - * / based on the operation parameter

def calculator(a, b, operation):
    if operation == '+':
        return a + b
    elif operation == '-':
        return a - b
    elif operation == '*':
        return a * b
    elif operation == '/':
        return a / b
    else:
        return "Invalid operation"

print(calculator(10, 5, '+')) 
```

```python
# write a function is_prime(n) that returns True if n is  a prime number OTHERWISE FALSE
# using a loop

def is_prime(n):
    if n <= 1:
        return False

    for i in range(2, n):
        if n % i == 0:
            return False

    return True

print(is_prime(7))  # True
```
