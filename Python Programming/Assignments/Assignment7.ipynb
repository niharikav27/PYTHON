```python
#Write a Python program to check if a specified element appears in a tuple of tuples.


t = (('Red', 'White', 'Blue'), ('Green', 'Pink', 'Purple'), ('Orange', 'Yellow', 'Lime'))

element = input("Enter element to search: ")       #White

found = False

for i in t:
    if element in i:
        found = True

print(found)
```

```python
#WAP to check the tuple contain any duplicate element
t = (1, 2, 3, 4, 5, 2)

if len(t) != len(set(t)):
    print("Tuple contains duplicate elements")
else:
    print("No duplicates found")
```

```python
#WAP to check if all the element of a tuple are in descending order or not.
t = (9, 7, 5, 3, 1)

if list(t) == sorted(t, reverse=True):
    print("Tuple is in descending order")
else:
    print("Tuple is NOT in descending order")
```

```python
#WAP to check if the elements in the first half of a tuple are sorted in ascending or not.
t = (1, 2, 3, 9, 5, 4)

mid = len(t) // 2
first_half = t[:mid]

if list(first_half) == sorted(first_half):
    print("First half is in ascending order")
else:
    print("First half is NOT in ascending order")
```

```python
#WAP to input names of n students and search a name
n = int(input("How many students? "))

names = []

for i in range(n):
    name = input(f"Enter name of student {i+1}: ")
    names.append(name)

t = tuple(names)

search = input("Enter name to be searched: ")

if search in t:
    print(search, "exists in tuple")
else:
    print(search, "does not exist in tuple")
```

```python
#Python program to find unique numbers in a given tuple
t = (1, 9, 1, 6, 3, 4, 5, 1, 1, 2, 5, 6, 7, 8, 9, 2)

unique = tuple(set(t))

print("Unique numbers:", unique)
```

```python
#Convert tuple of string values to tuple of integer values
t = (('333', '33'), ('1416', '55'))

new_tuple = tuple(tuple(int(x) for x in i) for i in t)

print("New tuple:", new_tuple)
```

```python
#write a program to calculate product of all numbers in a tuple
t = (4, 3, 2, 2, -1, 18)

product = 1

for i in t:
    product *= i

print("Product:", product)
```
